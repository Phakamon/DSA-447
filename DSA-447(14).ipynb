```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data
```

```python
X = data['X'].values
Y = data['Y'].values
```

```python
plt.scatter(X,Y,label = 'Scatter Plot')
plt.show()
```

```python
plt.scatter(X,Y)
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
print(mean_x)
print(mean_y)
```

```python
n = len(X)
print(n)
```

```python
#Least Squared Method
numerator = 0
denominator = 0

for i in range(n):
    numerator += (X[i] - mean_x) * (Y[i] - mean_y)
    denominator += (X[i] - mean_x) ** 2

m = numerator/denominator
print(m)
```

```python
c = mean_y -(m* mean_x)
c
```

```python
x = np.linspace(0,6)
y = m*x + c
```

```python
x = np.linspace(0,6)
y = m*x+c
```

```python
plt.plot(x,y, color = '#58b970')
plt.scatter(X,Y)
plt.xlabel('X Value')
plt.ylabel('Y Value')

plt.xlim(0,6)
plt.ylim(0,6)
plt.show()
```

```python
# Calculating R2 Score

ss_tot = 0
ss_res = 0

for i in range(n):
    y_pred = m * X[i] + c
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (y_pred - mean_y) ** 2

r2 = ss_res/ss_tot

print('R2 Score')
print(r2)
```
