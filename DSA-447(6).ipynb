```python
df3 = pd.read_excel('example.xlsx',  engine='openpyxl')
df3
```

```python
ohe = OneHotEncoder(sparse=False)
pd.DataFrame(ohe.fit_transform(df3[['Shape']]))
```

```python
oe = OrdinalEncoder(categories=[['first', 'second', 'third'], ['S', 'M', 'L', 'XL']])
pd.DataFrame(oe.fit_transform(df3[['Class','Size']]))
```

```python
df3
```

```python
one_hot_var = pd.DataFrame(ohe.fit_transform(df3[['Shape']]))
one_hot_var
```

```python
order_var = pd.DataFrame(oe.fit_transform(df3[['Class','Size']]))
order_var
```

```python
df = pd.concat([one_hot_var,order_var], axis='columns')
df
```
